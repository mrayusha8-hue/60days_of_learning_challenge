#Basic Plot Syntax

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

#Seaborn has 4 types of context('paper', notebook', 'talk', 'poster') which allows to override default parameters.
#Platte: Set1 - bright & high-contrast colors(9)
         Set2 - softer, more pastel-like colors(8)
         Set3 - mix of soft and bold,varied hues(12)

```python
#Scatter plot
sns.set_style("dark")
sns. set_context("notebook")
sns.scatterplot(data = penguins, x = "bill_length_mm", y ="bill_depth_mm", hue = "species",palette = "Dark2")
sns.despine(left = True)     #despine() removes the top and right spines from plots.(the borders around a plot)
```

```python
sns.scatterplot(penguins, x = "species", y = "body_mass_g", hue = "island", style = "sex", alpha = 0.5) #alpha, used as there are overlapping points, ,it balances the opacity of the points
```

```python
#Strip plot
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", dodge = True, jitter = True) #jitter used to spread randomly for more better readability
```

```python
#Swarm plot
sns.set_context("talk")
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "island") 
sns.despine()
```

```python
#Histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g",hue = "sex", multiple = "stack" ) # plots histogram of body_mass_g with hue categories stacked within each bin
```
