```python
#Univariate analysis

quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(quality_counts.index, quality_counts, color = 'orange')
plt.title('Count Plot of Wine Quality')
plt.xlabel('quality')
plt.ylabel('count')
plt.show()
```

This plot shows count of wine with its quality rate.
